HYDRAULIC ANALYSIS WITH LEAKS


```python
nodeset = d.getNodeEmitterCoeff()  
emitterCoefficient = numpy.arange(0,0.3,0.01)
```

```python
nodePressure1= []
nodePressure1= []
nodePressure2= []
nodePressure3= []
nodePressure4= []
nodePressure5= []
```

```python
for i in emitterCoefficient:
    nodeHead = []
    nodeset[5] = i
    epanet.openHydraulicAnalysis(d)
    d.setNodeEmitterCoeff(nodeset)                      # Sets the value of all nodes emitter coefficient
    d.solveCompleteHydraulics()
    
    d.setTimeHydraulicStep(3600)
    pressure1 = d.getNodePressure(1)
    pressure2 = d.getNodePressure(2)
    pressure3 = d.getNodePressure(3)
    pressure4 = d.getNodePressure(4)
    pressure5 = d.getNodePressure(5)

    #head = d.getNodeHydraulicHead(5)

    #nodePressure.append(i)
    nodePressure1.append(pressure1)
    nodePressure2.append(pressure2)
    nodePressure3.append(pressure3)
    nodePressure4.append(pressure4)
    nodePressure5.append(pressure5)

    #nodeHead.append(head)
    d.closeHydraulicAnalysis()
    print(nodeset)
```
